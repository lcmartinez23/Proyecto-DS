Para analizar la integración y el impacto de las Tecnologías de la Información y Comunicación (TIC) en el rendimiento estudiantil, basándonos en los datos disponibles en la base de datos situación académica al finalizar el año, realizaremos inicialmente la siguiente relacion:

1. Jornada Vs Situación Académica:
- Objetivo: Determinar si existe alguna relación entre la jornada escolar y la situación académica de los estudiantes al finalizar el año.
   - Tipo de Análisis: Inicialmente Comparación de frecuencias.
   - Gráfico Sugerido: barras para mostrar el número de estudiantes que Aprueban, reprueban, desertores, trasnferidos en cada jornada.

En consecuencia realizaremos un nuevo data_frame que contenga solamente los datos a relacionar.


```python
df_1= df[['JORNADA_NOMBRE','SITUACADE_NOMBRE']]
df_1.head()
```

Con este nuevo data_frame, realizaremos el conteo de la frecuencia de cada caterio de jornada y cada categoria de situacion académica.

```python
conteo_jornada = df['JORNADA_NOMBRE'].value_counts()
conteo_situacion = df['SITUACADE_NOMBRE'].value_counts()

print(conteo_jornada)
print(conteo_situacion)
```

Podemos graficar en barras la cantidad de estudiantes por jornada.

```python
conteo_jornada.plot(kind='bar')
plt.title('CANTIDAD DE ESTUDIANTES POR JORNADA')
plt.xlabel('JORNADA')
plt.ylabel('CANTIDAD DE ESTUDIANTES')
plt.show()
```

Podemos tambien graficar la cantidad de estudiantes por situación académica.

```python
conteo_situacion.plot(kind='barh')
plt.title('SITUACION ACADEMICA DE LOS ESTUDIANTES')
plt.xlabel('TIPO DE SITUACIÓN')
plt.ylabel('CANTIDAD DE ESTUDIANTES')
plt.show()
```

Ahora se hace necesario filtrar las situaciones academicas que se presentan en cada una de las jornadas, para lo cual filtramos el Data_frame df_1 para obtener solo las filas donde JORNADA_NOMBRE es igual a Mañana. Con el resultado de esta operación creamos un nuevo Data_frame llamado Conteo_df que contiene solo las filas que cumplen los criterios señalados en cada filtro.
con shape[0] estoy obteniendo el número de filas que cumple con el criterio

```python
Reprobados_mañana = df_1[(df_1['JORNADA_NOMBRE'] == 'Mañana') & (df_1['SITUACADE_NOMBRE'] == 'Reprobados')].shape[0]
desertores_mañana = df_1[(df_1['JORNADA_NOMBRE'] == 'Mañana') & (df_1['SITUACADE_NOMBRE'] == 'Desertores')].shape[0]
Trans_mañana = df_1[(df_1['JORNADA_NOMBRE'] == 'Mañana') & (df_1['SITUACADE_NOMBRE'] == 'Transferidos/Transladados')].shape[0]
Aprobados_mañana = df_1[(df_1['JORNADA_NOMBRE'] == 'Mañana') & (df_1['SITUACADE_NOMBRE'] == 'Aprobados')].shape[0]
conteo_df = pd.DataFrame({
    'Categoría': ['Reprobados', 'Desertores', 'Transferidos/Transladados', 'Aprobados'],
    'Mañana': [Reprobados_mañana, desertores_mañana, Trans_mañana, Aprobados_mañana]
})
conteo_df
```

Realizamos una grafica de barras con el data_frame conteo_df

```python
import matplotlib.pyplot as plt
plt.bar(conteo_df['Categoría'], conteo_df['Mañana'], color='skyblue')
plt.title('Conteo de Estudiantes por Categoría en la Jornada de Mañana')
plt.xlabel('Categoría')
plt.ylabel('Número de Estudiantes')

plt.show
```

Calculemos los porcentajes a los cuales corresponde cada categoria y grafiquemos.

```python
porcentajes = conteo_df['Mañana'] / conteo_df['Mañana'].sum() * 100
plt.figure(figsize=(4, 4))
plt.pie(porcentajes, labels=conteo_df['Categoría'], autopct='%1.1f%%', startangle=140)

plt.title('Porcentaje de Estudiantes por Categoría en la Jornada de Mañana')

plt.show()
```

```python
Reprobados_Unica = df_1[(df_1['JORNADA_NOMBRE'] == 'Unica') & (df_1['SITUACADE_NOMBRE'] == 'Reprobados')].shape[0]
desertores_Unica = df_1[(df_1['JORNADA_NOMBRE'] == 'Unica') & (df_1['SITUACADE_NOMBRE'] == 'Desertores')].shape[0]
Trans_Unica = df_1[(df_1['JORNADA_NOMBRE'] == 'Unica') & (df_1['SITUACADE_NOMBRE'] == 'Transferidos/Transladados')].shape[0]
Aprobados_Unica = df_1[(df_1['JORNADA_NOMBRE'] == 'Unica') & (df_1['SITUACADE_NOMBRE'] == 'Aprobados')].shape[0]
conteoU_df = pd.DataFrame({
    'Categoría': ['Reprobados', 'Desertores', 'Transferidos/Transladados', 'Aprobados'],
    'Unica': [Reprobados_Unica, desertores_Unica, Trans_Unica, Aprobados_Unica]
})

conteoU_df
```

```python
plt.bar(conteoU_df['Categoría'], conteoU_df['Unica'], color='green')
plt.title('Conteo de Estudiantes por Categoría en la Jornada de Unica')
plt.xlabel('Categoría')
plt.ylabel('Número de Estudiantes')
plt.show()
```

```python
porcentajes = conteoU_df['Unica'] / conteoU_df['Unica'].sum() * 100
plt.figure(figsize=(4, 4))
plt.pie(porcentajes, labels=conteoU_df['Categoría'], autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Estudiantes por Categoría en la Jornada de Unica')

plt.show()
```

```python
Reprobados_Completa = df_1[(df_1['JORNADA_NOMBRE'] == 'Completa') & (df_1['SITUACADE_NOMBRE'] == 'Reprobados')].shape[0]
desertores_Completa = df_1[(df_1['JORNADA_NOMBRE'] == 'Completa') & (df_1['SITUACADE_NOMBRE'] == 'Desertores')].shape[0]
Trans_Completa = df_1[(df_1['JORNADA_NOMBRE'] == 'Completa') & (df_1['SITUACADE_NOMBRE'] == 'Transferidos/Transladados')].shape[0]
Aprobados_Completa = df_1[(df_1['JORNADA_NOMBRE'] == 'Completa') & (df_1['SITUACADE_NOMBRE'] == 'Aprobados')].shape[0]
conteoC_df = pd.DataFrame({
    'Categoría': ['Reprobados', 'Desertores', 'Transferidos/Transladados', 'Aprobados'],
    'Completa': [Reprobados_Completa, desertores_Completa, Trans_Completa, Aprobados_Completa]
})
conteoC_df
plt.bar(conteoC_df['Categoría'], conteoC_df['Completa'], color='red')
plt.title('Conteo de Estudiantes por Categoría en la Jornada de Completa')
plt.xlabel('Categoría')
plt.ylabel('Número de Estudiantes')
plt.show()
porcentajes = conteoC_df['Completa'] / conteoC_df['Completa'].sum() * 100
plt.figure(figsize=(4, 4))
plt.pie(porcentajes, labels=conteoC_df['Categoría'], autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Estudiantes por Categoría en la Jornada de Completa')

plt.show()
```

```python
Reprobados_Tarde  = df_1[(df_1['JORNADA_NOMBRE'] == 'Tarde') & (df_1['SITUACADE_NOMBRE'] == 'Reprobados')].shape[0]
desertores_Tarde  = df_1[(df_1['JORNADA_NOMBRE'] == 'Tarde') & (df_1['SITUACADE_NOMBRE'] == 'Desertores')].shape[0]
Trans_Tarde = df_1[(df_1['JORNADA_NOMBRE'] == 'Tarde') & (df_1['SITUACADE_NOMBRE'] == 'Transferidos/Transladados')].shape[0]
Aprobados_Tarde  = df_1[(df_1['JORNADA_NOMBRE'] == 'Tarde') & (df_1['SITUACADE_NOMBRE'] == 'Aprobados')].shape[0]
conteoT_df = pd.DataFrame({
    'Categoría': ['Reprobados', 'Desertores', 'Transferidos/Transladados', 'Aprobados'],
    'Tarde': [Reprobados_Tarde , desertores_Tarde , Trans_Tarde , Aprobados_Tarde ]
})
conteoT_df
plt.bar(conteoT_df['Categoría'], conteoT_df['Tarde'], color='y')
plt.title('Conteo de Estudiantes por Categoría en la Jornada de Tarde')
plt.xlabel('Categoría')
plt.ylabel('Número de Estudiantes')
plt.show()
porcentajes = conteoT_df['Tarde'] / conteoT_df['Tarde'].sum() * 100
plt.figure(figsize=(4, 4))
plt.pie(porcentajes, labels=conteoT_df['Categoría'], autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Estudiantes por Categoría en la Jornada de Tarde')

plt.show()
```

```python
Reprobados_Finsem = df_1[(df_1['JORNADA_NOMBRE'] == 'Fin de Semana') & (df_1['SITUACADE_NOMBRE'] == 'Reprobados')].shape[0]
desertores_Finsem = df_1[(df_1['JORNADA_NOMBRE'] == 'Fin de Semana') & (df_1['SITUACADE_NOMBRE'] == 'Desertores')].shape[0]
Trans_Finsem = df_1[(df_1['JORNADA_NOMBRE'] == 'Fin de Semana') & (df_1['SITUACADE_NOMBRE'] == 'Transferidos/Transladados')].shape[0]
Aprobados_Finsem = df_1[(df_1['JORNADA_NOMBRE'] == 'Fin de Semana') & (df_1['SITUACADE_NOMBRE'] == 'Aprobados')].shape[0]
conteoF_df = pd.DataFrame({
    'Categoría': ['Reprobados', 'Desertores', 'Transferidos/Transladados', 'Aprobados'],
    'Fin de Semana': [Reprobados_Finsem, desertores_Finsem, Trans_Finsem, Aprobados_Finsem]
})
conteoF_df
plt.bar(conteoF_df['Categoría'], conteoF_df['Fin de Semana'], color='c')
plt.title('Conteo de Estudiantes por Categoría en la Jornada de Fin de Semana')
plt.xlabel('Categoría')
plt.ylabel('Número de Estudiantes')
plt.show()
porcentajes = conteoF_df['Fin de Semana'] / conteoF_df['Fin de Semana'].sum() * 100
plt.figure(figsize=(4, 4))
plt.pie(porcentajes, labels=conteoF_df['Categoría'], autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Estudiantes por Categoría en la Jornada de Fin de Semana')

plt.show()
```
